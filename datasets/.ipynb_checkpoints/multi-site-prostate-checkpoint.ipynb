# Preprocess of multi-site prostate data
data from https://liuquande.github.io/SAML/

```python
import sys
sys.path.append('../')
import SimpleITK as sitk
import numpy as np
import os
import glob
from PIL import Image
from utils import create_if_not
from preprocess import read_nii,copy_geometry,normalize_nonzero

data_dir = 'D:/Data/Multi-siteProstate/rawdata'
outputs_dir = 'D:/Data/Multi-siteProstate/outputs'
all_sites = os.listdir(data_dir)
print(all_sites)
create_if_not(outputs_dir)
```

```python
#convert from nifti to npy
for site in all_sites:
    site_path = outputs_dir+'/' + site
    create_if_not(site_path)
    volume_path = os.path.join(site_path,'volume')
    slices_path = os.path.join(site_path,'slices')
    create_if_not(volume_path)
    create_if_not(slices_path)
    seg_list = glob.glob(data_dir+'/'+site+'/*mentation.nii.gz')
    for idx,seg_path in enumerate(seg_list):
        img_path = seg_path.split('_')[0]+'.nii.gz'
        itkimg = sitk.ReadImage(img_path)
        itkseg = sitk.ReadImage(seg_path)
        np_img = sitk.GetArrayFromImage(itkimg)
        np_seg = sitk.GetArrayFromImage(itkseg)
        np_seg = (np_seg>0)*1
        np_img = np_img.astype(np.float32)
        np_seg = np_seg.astype(np.uint8)
        np_img= normalize_nonzero(np_img)
        #just use non-zero segmentation slices
        crop_mask = np.where(np_seg>0)
        crop_mask = crop_mask[0]
        #print(crop_mask,np.unique(np_seg))
        np_img = np_img[crop_mask[0]:crop_mask[-1]+1]
        np_seg = np_seg[crop_mask[0]:crop_mask[-1]+1]
        data = np.stack([np_img,np_seg])
        np.save(volume_path+'/volume{}.npy'.format(idx+1),data)
        #save 2d slices
        num_slices = np.shape(data)[1]
        for z in range(num_slices):
            img2d = data[0,z]
            seg2d = data[1,z]
            data2d = np.stack([img2d,seg2d])
            np.save(slices_path+'/slices{}_{}.npy'.format(idx+1,z+1),data2d)
    print('Site %s has %d volumes' % (site,idx+1))
    
```
